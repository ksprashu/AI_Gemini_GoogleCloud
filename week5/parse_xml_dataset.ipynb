```python
import re

def remove_curly_braces(text):
  """Removes text enclosed in double curly braces."""
  pattern = r"\{\{.*?\}\}"  # Non-greedy matching
  cleaned_text = re.sub(pattern, "", text)
  return cleaned_text


def remove_file_tags(text):
  """Removes text enclosed in [[File: ... ]] tags, including the tags."""
  pattern = r"\[\[File:.*?\]\]"  # Non-greedy matching
  cleaned_text = re.sub(pattern, "", text)
  return cleaned_text
```
